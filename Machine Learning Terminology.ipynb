## Supervised Machine Learning
### ML Terminology

ML systems learn how to combine input to produce useful predictions on never-before-seen data.
Let's explore fundamental machine learning terminology.

##### Labels
A label is the thing we're predicting—the $ y $ variable in simple linear regression. The label could be the future price of wheat, the kind of animal shown in a picture, the meaning of an audio clip, or just about anything

##### Feature
A feature is an input variable—the $x$ variable in simple linear regression. A simple machine learning project might use a single feature, while a more sophisticated machine learning project could use millions of features, specified as:

\begin{equation*}x_1,x_2,....x_N \end{equation*}

In the spam detector example, the features could include the following:

 - words in the email text
 - sender's address
 - time of day the email was sent
 - email contains the phrase "one weird trick."
 
**Examples**
An example is a particular instance of data, $x$. We break examples into two categories:

 - **Labeled examples**--> $(x^{(i)},y^{(i)})$
 - **Unlabeled examples**--> $(x^{(i)})$
 
##### Models

A model defines the relationship between features and label. For example, a spam detection model might associate certain features strongly with "spam". Let's highlight two phases of a model's life:

##### Training
__Training__ means creating or learning the model using the labeled examples and enable the model to gradually learn the relationships between features and label.

##### Inference
__Inference__ means applying the trained model to unlabeled examples. That is, you use the trained model to make useful predictions (y'). For example, during inference, you can predict _medianHouseValue_ for new unlabeled examples.

##### Scaling Techniques

 - MinMaxScaler:
\begin{equation*}
\frac{X_{i}-X_{min}}{X_{max}-X_{min}}
\end{equation*}

The range of **MinMaxScaller** is $(0,1)$. The Python code for this scaling is
<code>from sklearn.preprocessing import MinMaxScaler</code>

 - RobustScaler:
 
 The **RobustScaler** uses a similar method to the **Min-Max scaler**. However, it uses the interquartile range ($75\% - 25\%$) instead of the min-max, which makes it robust to **outliers**. It follows the following formula for each feature:
\begin{equation*}
\frac{X_{i}-Q_1}{IQR}=\frac{X_i-Q_1}{Q_3-Q_1}
\end{equation*}

The Python code for this scaling is <code>from sklearn.preprocessing import RobustScaler</code>

 - StandardScaler:
 \begin{equation*}
\frac{X_{i}-X_{mean}}{\sigma}
\end{equation*}

In general, **mean** of the distribution is $0$, and $68\%$ values lie in the range $(-1,1)$. Here $\sigma$ denotes the **Standard Deviation**. The Python code for this scaling is
<code>from sklearn.preprocessing import StandardScaler</code>. 

 - Normalizer
 
The normalizer scales each value by dividing each value by its magnitude in *n*-dimensional space for $n$ number of features.

Say our features were $x, y$ and $z$ Cartesian co-ordinates our scaled value for $x$ would be:
\begin{equation*}
\frac{x_i}{\sqrt{x_i+y_i+z_i}}
\end{equation*}

The Python code for this scaling is <code>from mpl_toolkits.mplot3d import Axes3D</code>.

##### fit(), transform(), fit_transform():

In the dataset, there are some missing values (<code>NaN</code>) which shoule be replaced with either <code>mean</code> or <code>median</code>. The Imputer has to calculate the <code>mean</code> or <code>median</code> first.

 - <code>fit()</code>: Calculate the <code>mean</code> or <code>median</code> internally.
 - <code>transform()</code>: Replace the <code>NaN</code> values with the newly calculated value (<code>mean</code> or <code>median</code>) internally, and return the new dataset with no <code>NaN</code> value.
 - <code>fit_transform()</code>: Calculate the <code>mean</code> or <code>median</code> value of a column and the rplace the <code>NaN</code> with it simultaneously.
 
During training a dataset, we use the <code>fit()</code> method to claculate <code>mean</code> or <code>median</code> values and save internally. However, in case of test data, we don't need to calculate these values again. The thing we need just replace the <code>NaN</code> values of test data with the previoulsy calculated values using the <code>transform()</code> method.

To put it simply, we can use the <code>fit_transform()</code> method on the training set, as we’ll need to both fit and transform the data, and we can use the <code>fit()</code> method on the training dataset to get the value, and later <code>transform()</code> test data with it.

##### Random_state:

While using <code>train_test_split</code> of <code>sklearn</code> to divide the entire dataset into <code>train_set</code> and <code>test_set</code>, we have to specify the any integer value for <code>random_state</code> $(0, \forall n\in \mathbb N)$. If we don't mention the **random_state** in the code, then whenever we execute your code a new random value is generated and the train and test datasets would have different values each time. Therefore, everytime, we will get a different accuracy for our trained model on the test data. It doesn't matter how many times we execute the code and what the value of <code>random_state</code> is, it will provide us with the same <code>train/test_set</code> each time.

#### Linear Regression
A **regression** model predicts continuous values. For example, regression models make predictions that answer questions like the following:

  - What is the value of a house in California?

  - What is the probability that a user will click on this ad?

The cost function for $m$ training examples, 
\begin{equation*}
J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}\left(h_{\theta}(x^{(i)})-y^{(i)}\right)^2
\end{equation*}

In the above equation, $h_{\theta}(x^{(i)})=\theta_0+\theta_1x$, where $\theta_0$ is the <code>bias</code> and $\theta_1$ is the <code>slope</code>.

Another name of the cost function ($J_{\theta}(x)$) is called the Mean Squared Error (**MSE**).

**Gradient Descent Algorithm**

\begin{figure}
  \includegraphics[width=\linewidth]{gradient-descent.PNG}
\end{figure}

\begin{equation*}\theta_j = \theta_j-\alpha \frac{\partial}{\partial \theta_j}J(\theta_o,\theta_1) \end{equation*}
Where $j=0$ and $j=1$.

**Derivation of $\theta$ using OLS method**

\begin{equation*}
  J = \frac{1}{2m}\sum{\left(h_{\theta}(x^{(i)})-y^{(i)}\right)^2}=\frac{1}{2m}\sum{\left(\theta^{T}X-y^{(i)}\right)^2}=\frac{1}{2m}||\theta X - y||^2 = \frac{1}{2m}\left(\theta X - y \right)^T \left(\theta X - y \right)
\end{equation*}
\begin{equation*}
  =\frac{1}{2m}\left(\theta^T X^T - y^T \right)\left(\theta X - y \right) = \frac{1}{2m}\left(\theta^T X^T \theta X - y^T \theta X - \theta^T X^T y +y^T y \right)
\end{equation*}


\begin{equation*}
\frac{\partial J}{\partial \theta} = \frac{\partial}{\partial \theta}\left[\frac{1}{2m}\left(\theta^T X^T \theta X - y^T \theta X - \theta^T X^T y +y^T y \right) \right] = \frac{1}{2m} 2 \left(\theta X^TX - X^Ty - X^Ty+0 \right)
\end{equation*}

To minimize the cost function $J$, consider $\frac{\partial J}{\partial \theta}=0$, where $X^TX$ is invertible

\begin{equation*}
\theta X^TX-X^Ty=0 \Rightarrow \theta \left(X^TX \right) = X^Ty \Rightarrow \theta = \left(X^TX \right)X^Ty
\end{equation*}

\begin{equation*}
\therefore \theta = \left(X^TX \right)^{-1}X^Ty
\end{equation*}

**The learning rate $\alpha$ (also sometimes called step size):**

The ideal learning rate in one-dimension is $\frac{1}{J(\theta)^n}$ (the inverse of the second derivative of $J(\theta)$ at $\theta$).

The ideal learning rate for **2** or more dimensions is the inverse of the Hessian (matrix of second partial derivatives).

For example the Hessian Matrix of $J(\theta_0,\theta_1,\theta_2,....\theta_n)$ is defined below:
\begin{equation*}
H J(\theta_0,\theta_1,\theta_2,....\theta_n) = \begin{pmatrix} % you can specify any environment according to your choice  
\frac{\partial^2 J}{\partial \theta_0^{2}} & \frac{\partial^2 J}{\partial \theta_0 \partial \theta_1} & \frac{\partial^2 J}{\partial \theta_0 \partial \theta_2} & \cdots & \frac{\partial^2 J}{\partial \theta_0 \partial \theta_n} \\  
\frac{\partial^2 J}{\partial \theta_1 \partial \theta_0} & \frac{\partial^2 J}{\partial \theta_1^{2}} & \frac{\partial^2 J}{\partial \theta_1 \partial \theta_2} & \cdots & \frac{\partial^2 J}{\partial \theta_1 \partial \theta_n} \\ % for more than one digit, we need to enclose the underscore value inside the curly brackets.  
\frac{\partial^2 J}{\partial \theta_2 \partial \theta_0} & \frac{\partial^2 J}{\partial \theta_2 \partial \theta_1} & \frac{\partial^2 J}{\partial \theta_2^{2}} &  \cdots & \frac{\partial^2 J}{\partial \theta_2 \partial \theta_n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\  
\frac{\partial^2 J}{\partial \theta_n \partial \theta_0} & \frac{\partial^2 J}{\partial \theta_n \partial \theta_1} & \frac{\partial^2 J}{\partial \theta_n \partial \theta_2} & \cdots & \frac{\partial^2 J}{\partial \theta_{n}^2}   
\end{pmatrix}
\end{equation*}

Using the concept of Newton's method,
\begin{equation*}
\theta_{n+1}=\theta_n-\frac{J^{'}(\theta_n)}{J^{''}(\theta_n)}=\theta_n-\left[J^{''}(\theta_n)\right]^{-1}*J^{'}(\theta_n)
\end{equation*}

Since in Hessian Matrix, the second order derivative of $n$ variable is used, then the gradient descent can be defined as 
\begin{equation*}
\theta_{n+1}=\theta_n-\left[HJ(\theta_n)\right]^{-1}*\nabla J(\theta_n)
\end{equation*}
 - If $\alpha$ is very small, then the gradient descent can be very slow
 - If $\alpha$ is very large, then the gradient descent can overshoot the minimum. It may fail to converge, or even diverge

**Derivation of $\theta$ using Newton's Method:**
\begin{equation*}
H J(\theta) = \frac{\partial}{\partial \theta}\left(\nabla_{\theta}J \right) = \frac{\partial}{\partial \theta} \left(\frac{\partial J}{\partial \theta} \right)= \frac{\partial}{\partial \theta}\left(2X^T\theta X - 2X^Ty \right)=2X^TX
\end{equation*}

Now, \begin{equation*}
\theta \leftarrow \theta - \left(H J(\theta) \right)^{-1}
=\theta - \left(2X^TX \right)^{-1}\left(2(X^X \theta - X^Ty)\right)=\theta - \left(X^TX \right)^{-1}(X^TX)\theta + \left(X^TX \right)^{-1}X^Ty \\
\therefore \theta = \left(X^TX \right)^{-1}X^Ty
\end{equation*}
Which is as same as the derivation using **OLS method**, and it is clear that the **convergence** happens much faster than that of the **gradient descent** method. Everyone should bear in mind that Newton's method is worthy to be applied only for small number of features.

 - **Stochastic Gradient Descent:**
 
Stochastic gradient descent (SGD) takes this idea to the extreme--it uses only a single example (a batch size of 1) per iteration. Given enough iterations, SGD works but is very noisy. The term "**stochastic**" indicates that the one example comprising each batch is chosen at random.

 - **Mini-Batch Gradient Descent:**
 
Mini-batch stochastic gradient descent (**mini-batch SGD**) is a compromise between full-batch iteration and SGD. A mini-batch is typically between 10 and 1,000 examples, chosen at random. Mini-batch SGD reduces the amount of noise in SGD but is still more efficient than full-batch.

#### Polynomial Regression

Polynomial features are often created when we want to include the notion that there exists a **nonlinear relationship** between the features and the target.They are mostly used to add complexity to linear models with little features, or when we suspect the effect of one feature is dependent on another feature.

Before handling missing values, you need to decide if you want to use polynomial features or not. If you for example replace all the missing values by 0, all the cross-products using this feature will be $0$. Moreover, if you don’t replace missing values (**NaN**), creating polynomial features will raise a value error in the <code>fit_transform</code> phase, since the input should be finite.

 - <code>degree</code>: Determines the maximum degree of the polynomial.
 - <code>interaction_only</code>: Suppose, there are three features ($f_1,f_2,f_3$), the the <code>interaction_only</code> parameter will produce some combined features $p_1 \rightarrow f_1.f_2, p_2 \rightarrow f_1.f_3, p_2 \rightarrow f_2.f_3, p_4 \rightarrow f_1.f_2.f_3$. Note that our original features are also included in the output and we slice off the new features to add to our data later.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)

# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.show()
```

```python
import operator

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)

# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]

polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
print(rmse)
print(r2)

plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

#### Binary Classification

**Sigmoid Function**
A function that maps logistic or multinomial regression output (log odds) to probabilities, returning a value between 0 and 1. The sigmoid function has the following formula:

\begin{equation*}
g(z)=\frac{1}{1+e^{-z}}
\end{equation*}
where  in logistic regression problems is simply: $z=\theta_0+\theta_1 x_1+\theta_2 x_2+.....+\theta_n x_n$

\begin{figure}
 \includegraphics[width=\linewidth]{sigmoid.PNG}
\end{figure}

Now, we can define conditional probabalities for 2 labels (0 and 1) for $i^{th}$ observation as:
\begin{equation*}
 P\left(y^{(i)}=1|x^{(i)};\theta \right)=h\left(x^{(i)} \right)\\
 P\left(y^{(i)}=0|x^{(i)};\theta \right)=1-h\left(x^{(i)} \right)
\end{equation*}

Now, it can be written more compactly (Bernoulli's disribution) as:
\begin{equation*}
P\left(y^{(i)}|x^{(i)};\theta \right)=\left(h\left(x^{(i)} \right)\right)^{y^{(i)}}\left(1-h\left(x^{(i)} \right)\right)^{1-y^{(i)}}
\end{equation*}

Now, we define another term, likelihood of **parameters** as:

\begin{equation*}
 L(\theta)=\prod_{i=1}^{m}P\left(y^{(i)}|x^{(i)};\theta \right)=\prod_{i=1}^{m}\left(h\left(x^{(i)} \right)\right)^{y^{(i)}}\left(1-h\left(x^{(i)} \right)\right)^{1-y^{(i)}}
\end{equation*}

For Easier calculation, we take **log likelihood**:

\begin{equation*}
 l(\theta)=\log \left(L(\theta) \right)=\sum_{i=1}^m \left[ y^{(i)}\log \left(h\left(x^{(i)}\right)\right) + \left(1-y^{(i)}\right)\log \left(1-h\left(x^{(i)}\right)\right) \right]
\end{equation*}

The cost function for logistic regression is proportional to inverse of likelihood of parameters. Hence, we can obtain an expression for cost function, $J$ using log likelihood equation as:

\begin{equation*}
J(\theta) \propto \log \left [L(\theta)\right]^{-1}=\frac{1}{m}\log \left [L(\theta)\right]^{-1}=\frac{1}{m} \log\left [\prod_{i=1}^{m}P\left(y^{(i)}|x^{(i)};\theta \right) \right]^{-1}\\
=\frac{1}{m}\log\left [\prod_{i=1}^{m}\left(h\left(x^{(i)} \right)\right)^{y^{(i)}}\left(1-h\left(x^{(i)} \right)\right)^{1-y^{(i)}} \right]^{-1}\\
=-\frac{1}{m}\sum_{i=1}^m\left[ y^{(i)}\log \left(h\left(x^{(i)}\right)\right) + \left(1-y^{(i)}\right)\log \left(1-h\left(x^{(i)}\right)\right) \right]\\
\end{equation*}
In the above illustration, $m$ is the total number of features (**Total columns**)

Now, cost function for a single feature $x$ can be simplified as:
\begin{equation*}
    Cost(h(x))= 
\begin{cases}
    -\log(h(x))\, & \text{if } y= 1\\
    -\log(1-h(x))\, & \text{if } y= 0
\end{cases}
\end{equation*}

##### Gradient Descent

\begin{equation*}
 \frac{\partial}{\partial \theta_j}\left[\log\left(h\left(x^{(i)}\right)\right)\right]=\frac{\partial}{\partial\theta_j}\left[\log\left(\frac{1}{1+e^{-\theta x^{(i)}}}\right)\right]=\frac{\partial}{\partial\theta_j}\left[-\log\left(1+e^{-\theta x^{(i)}}\right)\right]
 =-\frac{e^{-\theta x^{(i)}}}{1+e^{-\theta x^{(i)}}}\left(-x_j^{(i)}\right)=x_j^{(i)}\left(1-h\left(x^{(i)} \right)\right)\\
 \frac{\partial}{\partial \theta_j}\left[\log\left(1-h\left(x^{(i)}\right)\right)\right]=\frac{\partial}{\partial\theta_j}\left[\log\left(1-\frac{1}{1+e^{-\theta x^{(i)}}}\right)\right]=\frac{\partial}{\partial \theta_j}\left[\log\left(\frac{e^{-\theta x^{(i)}}}{1+e^{-\theta x^{(i)}}}\right)\right]=\frac{\partial}{\partial \theta_j}\left[\log\left(\frac{1}{1+e^{\theta x^{(i)}}}\right)\right]=\frac{\partial}{\partial\theta_j}\left[-\log\left(1+e^{\theta x^{(i)}}\right)\right]\\=-\frac{e^{\theta x^{(i)}}}{1+e^{\theta x^{(i)}}}\left(x_j^{(i)}\right)
 =-\frac{1}{1+e^{-\theta x^{(i)}}}\left(x_j^{(i)}\right)=\left(-x_j^{(i)}\right)h\left(x^{(i)}\right)
\end{equation*}

Now,
\begin{equation*}
 \frac{\partial J}{\partial \theta_j}=-\sum_{i=1}^{m}\left[y^{(i)}x_j^{(i)}\left(1-h\left(x^{(i)} \right)\right)+\left(1-y^{(i)}\right)(-x_j^{(i)})h\left(x^{(i)}\right)\right]=\sum_{i=1}^{m}\left[h \left(x^{(i)}\right)-y^{(i)}\right]x_{j}^{(i)}=X^{T}\left[h \left(x\right)-y\right]
\end{equation*}
Now, the gradual update of the gradient descent can be defined as:
\begin{equation*}
 \theta_{j} := \theta_{j} - \alpha\frac{\partial J}{\partial \theta_j}\\
 \therefore \theta_{j} := \theta_{j} - \alpha \sum_{i=1}^{m}\left[h \left(x^{(i)}\right)-y^{(i)}\right]x_{j}^{(i)}
\end{equation*}

#### Multi-class Classification (Softmax)
In the **softmax regression** setting, we are interested in multi-class classification (as opposed to only binary classification), and so the label $y$ can take on $K$ different values, rather than only two. Thus, in our training set $\left{\left(x^{(1)},y^{(1)}\right),\left(x^{(2)},y^{(2)}\right),…,\left(x^{(m)},y^{(m)}\right)\right}$, we now have that $y^{(i)}\epsilon\left{1,2,…,K\right}$

\begin{figure}
 \includegraphics[width=\linewidth]{softmax.PNG}
\end{figure}

Given a test input $x$, we want our hypothesis to estimate the probability that $P(y=k|x)$ for each value of $k=1,2,…,K$. I.e., we want to estimate the probability of the class label taking on each of the $K$ different possible values. Thus, our hypothesis will output a $K$-dimensional vector (whose elements sum to 1) giving us our $K$ estimated probabilities. Concretely, our hypothesis $h_{\theta}(x)$ takes the form:

\begin{equation*}
 h_{\theta}(x) = \begin{pmatrix}
 P(y=1|x)\\
 P(y=2|x)\\
 P(y=3|x)\\
 \vdots\\
 P(y=K|x)
\end{pmatrix}=\frac{1}{\sum_{j=1}^{K}e^{\theta^{(j)T}x}}\times \begin{pmatrix}e^{\theta^{(1)T}x}\\ e^{\theta^{(2)T}x}\\ e^{\theta^{(3)T}x}\\ 
\vdots\\ e^{\theta^{(K)T}x}\end{pmatrix}
\end{equation*}

The above formula, in **Softmax Regression**, can be simplifies as:
\begin{equation*}
P(y^{(i)}=k|x^{(i)};θ)=h_{\theta}(x)=\frac{e^{\theta^{(j)T}x^{(i)}}}{\sum_{j=1}^{K}e^{\theta^{(j)T}x^{(i)}}}
\end{equation*}

Here $θ^{(1)},θ^{(2)},θ^{(3)}…,θ^{(K)}\epsilon\Re^n$ are the parameters of our model. Notice that the term $\frac{1}{\sum_{j=1}^{K}e^{\theta^{(j)T}x}}$ normalizes the distribution, so that it sums to one.

For convenience, we will also write $\theta$ to denote all the parameters of our model. When you implement softmax regression, it is usually convenient to represent $\theta$ as a $n \times K$ matrix obtained by concatenating $θ^{(1)},θ^{(2)},…,θ^{(K)}$ into columns, so that

\begin{equation*}
\theta = 
 \begin{pmatrix}
 \vdots & \vdots & \vdots & \vdots\\
 \theta^{(1)} & \theta^{(2)} &\cdots& \theta^{(K)}\\
 \vdots & \vdots & \vdots & \vdots 
 \end{pmatrix}
\end{equation*}

##### Cost Function

\begin{equation*}
 J(\theta)=-\frac{1}{m}\left[\sum_{i=1}^{m}\sum_{k=1}^{K}1\big\{y^{(i)}=k\big\} \log\left(\frac{e^{\theta^{(j)T}x^{(i)}}}{\sum_{j=1}^{K}e^{\theta^{(j)T}x^{(i)}}}\right)\right]
\end{equation*}

##### Gradient Descent

The gadient of the cost function is
\begin{equation*}
 \nabla_{\theta^{k}}J(\theta)=-\sum_{i=1}^{m}\left[1\big\{y^{(i)}=k \big\}-P(y^{(i)}=k|x^{(i)};\theta)\right]
\end{equation*}

##### Activation Function

A function (for example, **ReLU** or **sigmoid**) that takes in the weighted sum of all of the inputs from the previous layer and then generates and passes an output value (typically nonlinear) to the next layer. There are various types of **Activation Function** which are mentioned below
 - **Sigmoid/Logistic Activation:** Since the range of the **Sigmoid function** is always $(0, 1)$ and the range of the probability of occuring/existing anything is also $(0, 1)$, sigmoid is the right choice.
 
 \begin{equation*}
Sigmoid(z)=g(z)=\frac{1}{1+e^{-z}}
\end{equation*}

 - **Softmax Activation:** The softmax function is a more generalized logistic activation function which is used for **multiclass classification**.
 - **Tanh or Hyperbolic Activation:** $tanh$ (known as **scaled sigmoid** functions) is also like logistic sigmoid but better. Because the range of **Sigmoid** is $(0, 1)$, whereas the range of the **tanh function** is $(-1, 1)$. Surprisingly, <code>tanh</code> is also **sigmoidal (s - shaped)**.
 
 \begin{equation*}
 tanh(x) = 2Sigmoid(2x)-1 = \frac{2}{1+e^{-2x}}-1 = \frac{1-e^{-2x}}{1+e^{-2x}}=\frac{e^x-e^{-x}}{e^x+e^{-x}}=\frac{\frac{e^x-e^{-x}}{2}}{\frac{e^x+e^{-x}}{2}}=\frac{\sinh(x)}{\cosh(x)}
 \end{equation*}

   **Note:** *Both tanh and logistic sigmoid activation functions are used in feed-forward nets.* 

 - **ArcTan Activation:**
 
 \begin{equation*}
 f(x) = tan^{-1}(x)
 \end{equation*}

 - **Binary Activation:**
   \begin{equation*}
   g(z)= 
   \begin{cases}
   0 & \text{for } z\lt 0\\
   1 & \text{for } z\geq 0 
   \end{cases}
   \end{equation*}

 - **ReLU Activation:**
   - Simple and efficient: It is said to have 6 times improvement in convergence from tanh function 
   - Range: $[0, \infty)$
   - Avoids Vanishing Gradient problem
   - Can only be used within hidden layer in Neural Network model
   \begin{equation*}
   g(z)=max(0, z) = 
   \begin{cases}
   0 & \text{for } z\lt 0\\
   z & \text{for } z\geq 0 
   \end{cases}
   \end{equation*}

 - **Parametric ReLU (PReLU):**
    \begin{equation*}
   g(z)= 
   \begin{cases}
   \alpha z & \text{for } z\lt 0\\

#### Regularization (Regularized Regression)
Regularization is a very important concept that is used to avoid **overfitting** of the data especially when the trained and tested data are much varying.

Regularization is implemented by adding a **“penalty”** term to the best fit derived from the trained data, in order to achieve a **lesser variance** with the tested data and also restricts the influence of predictor variables over the output variable by **compressing/shrinking** their coefficients.
##### Ridge Regularization

\begin{figure}
 \includegraphics[width=\linewidth]{ridge.PNG}
\end{figure}

 - Performs $L_2$ regularization, i.e. adds penalty equivalent to square of the magnitude of coefficients
 - Minimization objective = LS Obj + $\lambda$ * (sum of square of coefficients)

\begin{equation*}
y = \theta_0 + \theta_1x_1 + \theta_2x_2 + ··· \theta_nx_n + \lambda (Slope)^2
\end{equation*}

The penalty term for ridge regression is $\lambda (slope)^2$, where $\lambda$ denotes the degree of deflection (<code>regularization rate</code>) from the original curve by restricting the coefficients of predictor variables but never makes them **zero**.

The **cost function** for $m$ training examples and $n$ features, 
\begin{equation*}
J(\theta)=\frac{1}{2m}\left[\sum_{i=1}^{m}\left(h_{\theta}(x^{(i)})-y^{(i)}\right)^2 +{\lambda}\sum_{j=1}^{n}\theta_{j}^2 \right]
\end{equation*}
In the above formula of **cost function**, 

 - When $\lambda \rightarrow 0$, then the data has <code>high variance</code>, and <code>overfit</code> the model. Then the model will learn too much about the particularities of the training data, and won't be able to generalize to new data
 - When $\lambda \rightarrow \infty$, then the data has <code>high bias</code>, and <code>underfit</code> the model. Then the model won't learn enough about the training data to make useful predictions.
 
**Gradient Descent for Ridge Regression**

\begin{equation*}
 \theta_{j} := \theta_{j} - \alpha\frac{\partial J}{\partial \theta_j}\\
 \theta_{0} := \theta_{0} - \frac{\alpha}{m}\left[ \sum_{i=1}^{m}\left(h \left(x^{(i)}\right)-y^{(i)}\right)x_{0}^{(i)}\right]\\
 \therefore \theta_{j} := \theta_{j} - \frac{\alpha}{m}\left[ \sum_{i=1}^{m}\left(h \left(x^{(i)}\right)-y^{(i)}\right)x_{j}^{(i)} + \lambda \theta_{j}\right]
\end{equation*}

##### Lasso (Least Absolute Shrinkage and Selection Operator ) Regularization

 - Performs $L_1$ regularization, i.e. adds penalty equivalent to absolute value of the magnitude of coefficients
 - Minimization objective = LS Obj + $\lambda$ * (sum of absolute value of coefficients)

\begin{equation*}
y = \theta_0 + \theta_1x_1 + \theta_2x_2 + ··· \theta_nx_n + \lambda |Slope|
\end{equation*}

Lesso regression can even eliminate the variables by making their coefficients to zero thus removing the variables that have high covariance with other predictor variables.

The **cost function** for $m$ training examples and $n$ features, 
\begin{equation*}
J(\theta)=\frac{1}{2m}\left[\sum_{i=1}^{m}\left(h_{\theta}(x^{(i)})-y^{(i)}\right)^2 +{\lambda}\sum_{j=1}^{n}|\theta_{j}| \right]
\end{equation*}

##### Epochs, Iterations & Batch size

Assuming there is a nice music track that comes with with lyrics and made of **six (6) verses** that we want to learn. After playing the full music track (**Epoch**) and going through each of the **6 verses (Iterations)** for the first time, there is **no guarrantee** we will be able to sing this song on our own without refering to the lyrics or **replaying** the song. Thus, we may need to replay the song for a sufficient number of times(**Multiple Epochs**) until perfection or confident enough to sing on our own or even recite any verse from any parts of the song (**High learning acuracy**).

In the similar way, if our training dataset has $1000$ records, we could decide to split it into $10$ batches (<code>100 records per batch — Batch size of 100</code>). Thus, **10 steps (iterations=10)** will be required to complete one learning cycle. There is no guarantee that the **Gradient Descent** will be globally optimized or reach its best optimization by the end of the first optimization cycle(**Epoch**). Because of this, **more Epochs** are often needed to achieve an ideal or high model accuracy. There is no set number of Epochs for optimizing a particular learning algorithm.

Just one Epoch can result to underfitting. However, **too many Epochs** after reaching **global minimum** can cause learning model to **overfit**. Ideally, the right number of epoch is one that results to the highest accuracy of the learning model.

##### Bias and Variance

###### What is bias?
Bias is how far are the **predicted values** from the **actual values**. If the average predicted values are far off from the actual values then the bias is high.

High bias causes algorithm to miss relevant relationship between input and output variable. When a model has a high bias then it implies that the model is too simple and does not capture the complexity of data thus underfitting the data.

###### What is variance?
Variance occurs when the model performs good on the trained dataset but does not do well on the <code>unseen dataset</code> (**test dataset or validation dataset**). Variance tells us how scattered are the predicted value from the actual value.

<code>High variance</code> causes **overfitting** that implies that the algorithm models random noise present in the training data.

\begin{figure}
 \includegraphics[width=\linewidth]{bias_variance.PNG}[scale=1.6]
\end{figure}

From the above figure, it can be decided that
 - **High Bias Low Variance:** Models are consistent but inaccurate on average
 - **High Bias High Variance:** Models are inaccurate and also inconsistent on average
 - **Low Bias Low Variance:** Models are accurate and consistent on averages. We strive for this in our model
 - **Low Bias High variance:** Models are somewhat accurate but inconsistent on averages. A small change in the data can cause a large error.
 
**Is there a way to find when we have a high bias or a high variance?**

<code>High Bias</code> can be identified when we have
 - High training error
 - Validation error or test error is same as training error
 
<code>High Variance</code> can be identified when
 - Low training error
 - High validation error ot high test error
 
**How do we fix high bias or high variance in the data set?**

<code>High bias</code> is due to a simple model and we also see a <code>high training error</code>. To fix that we can do following things
 - Add more input features
 - Add more complexity by introducing polynomial features
 - Decrease Regularization term (**Decrease** $\lambda$)

<code>High variance</code> is due to a model that tries to fit most of the training dataset points and hence gets more complex. However, it coud not do well in case of <code>test/CV data</code>. To resolve this issue we can do the the following things
 - Getting more training data
 - Reduce input features
 - Increase Regularization term (**Increase** $\lambda$)

 - **Agglomerative:** 
Bottom up approach. Start with many small clusters and merge them together to create bigger clusters.

### K-means Clustering
In case of k-means clutsering, the new centroid is determined using the mean of the points in their respective clsuter.

In addition, the K-means clustering relies on the **Euclidean Distance** of the points from the centroid. For example, if $(2,5)$  be the centroid and $(6,2)$  be a point in the cluster, then the **Euclidean Distance** between the above-mentioned points is $\sqrt{(2-6)^2+(5-2)^2}=\sqrt{(-4)^2+3^2=}=\sqrt{16+9}=\sqrt{25}=5$

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

### K-median Clustering
In statistics and data mining, k-medians clustering is a cluster analysis algorithm. It is a variation of k-means clustering where instead of calculating the mean for each cluster to determine its centroid, one instead calculates the median. 

The K-median clustering relies on the **Manhattan distance (rectilinear distance)** of the points from the centroid. For example: For example, if $(2,5)$ be the centroid the k-median distance between $(6,2)$ be a point in the cluster, then the **rectilinear distance** between the above-mentioned points is $|2-6|+|5-2|=|-4|+|3|=4+3=7$
